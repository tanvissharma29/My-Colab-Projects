**Testing and deployment of model**

PIL : Pillow Python Imaging Library
Both numpy and PIL are used for image preprocessing
Gradio is used for making machine learning model, prototype
CV2 is open source computer vision library

```python
!pip install -q gradio opencv-python-headless

import cv2
import numpy as np
from PIL import Image
import gradio as gr

# Simple Fire/Smoke detection using color thresholding
def detect_fire_smoke_demo(image):
    # Convert PIL to OpenCV BGR
    img = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    output = img.copy()

    # Convert to HSV color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Fire-like colors (red/orange/yellow)
    lower_fire = np.array([0, 120, 180])
    upper_fire = np.array([25, 255, 255])
    mask_fire = cv2.inRange(hsv, lower_fire, upper_fire)

    # Smoke-like colors (gray)
    lower_smoke = np.array([0, 0, 50])
    upper_smoke = np.array([180, 50, 200])
    mask_smoke = cv2.inRange(hsv, lower_smoke, upper_smoke)

    # Find contours for Fire
    contours_fire, _ = cv2.findContours(mask_fire, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours_fire:
        if cv2.contourArea(cnt) > 500:  # filter small noise
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(output, (x, y), (x+w, y+h), (0,0,255), 2)
            cv2.putText(output, "Fire", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)

    # Find contours for Smoke
    contours_smoke, _ = cv2.findContours(mask_smoke, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours_smoke:
        if cv2.contourArea(cnt) > 500:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(output, (x, y), (x+w, y+h), (255,0,0), 2)
            cv2.putText(output, "Smoke", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,0,0), 2)

    # Convert back to RGB for display
    output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
    return Image.fromarray(output)

# Gradio interface
demo = gr.Interface(
    fn=detect_fire_smoke_demo,
    inputs=gr.Image(type="pil", label="Upload Image"),
    outputs=gr.Image(label="Detection Result"),
    title="🔥 Fire & Smoke Detection Demo",
    description="This is a demo version using color detection for Fire and Smoke."
)

demo.launch(share=True, inline=True)
```
